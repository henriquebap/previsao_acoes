# ğŸ“Š Entendendo as MÃ©tricas de AvaliaÃ§Ã£o

Este notebook explica **em detalhes** cada mÃ©trica usada para avaliar modelos de previsÃ£o.

## ğŸ“š MÃ©tricas que vamos aprender

| MÃ©trica | Nome Completo | O que mede |
|---------|---------------|------------|
| **RMSE** | Root Mean Square Error | Erro mÃ©dio em dÃ³lares |
| **MAE** | Mean Absolute Error | Erro absoluto mÃ©dio |
| **MAPE** | Mean Absolute Percentage Error | Erro percentual mÃ©dio |
| **RÂ²** | Coeficiente de DeterminaÃ§Ã£o | Quanto o modelo explica |
| **Dir. Acc** | Directional Accuracy | Acerto na direÃ§Ã£o |

---

## Por que mÃºltiplas mÃ©tricas?

Cada mÃ©trica conta uma **parte da histÃ³ria**:
- RMSE/MAE: Quanto erramos em valor absoluto
- MAPE: Quanto erramos em porcentagem
- RÂ²: QuÃ£o bem o modelo se ajusta aos dados
- AcurÃ¡cia Direcional: Acertamos se sobe ou desce?


```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo
actual = np.array([100, 102, 98, 105, 110, 108, 115, 120, 118, 125])
predicted = np.array([101, 100, 99, 103, 112, 106, 118, 117, 120, 122])

print("ğŸ“Š Dados de Exemplo")
print("=" * 40)
for i, (a, p) in enumerate(zip(actual, predicted), 1):
    erro = p - a
    print(f"Dia {i}: Real=${a:.0f} | Previsto=${p:.0f} | Erro=${erro:+.0f}")
```

## 1ï¸âƒ£ RMSE - Root Mean Square Error

**FÃ³rmula**: $RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$

**O que significa**:
- Erro mÃ©dio em **dÃ³lares**
- Penaliza erros grandes mais do que erros pequenos
- RMSE menor = modelo melhor


```python
# CÃ¡lculo do RMSE passo a passo
errors = predicted - actual
squared_errors = errors ** 2
mean_squared_error = np.mean(squared_errors)
rmse = np.sqrt(mean_squared_error)

print("ğŸ“Š CÃ¡lculo do RMSE:")
print(f"   1. Erros: {errors}")
print(f"   2. ErrosÂ²: {squared_errors}")
print(f"   3. MÃ©dia dos errosÂ²: {mean_squared_error:.2f}")
print(f"   4. RMSE = âˆš{mean_squared_error:.2f} = ${rmse:.2f}")
print(f"\nâœ… RMSE = ${rmse:.2f}")
print("   InterpretaÃ§Ã£o: Em mÃ©dia, erramos ${:.2f} por previsÃ£o".format(rmse))
```

## 2ï¸âƒ£ MAE - Mean Absolute Error

**FÃ³rmula**: $MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$

**DiferenÃ§a do RMSE**: MAE nÃ£o eleva ao quadrado, entÃ£o trata todos os erros igualmente.


```python
# MAE
mae = np.mean(np.abs(errors))
print(f"âœ… MAE = ${mae:.2f}")

# MAPE
mape = np.mean(np.abs(errors / actual)) * 100
print(f"âœ… MAPE = {mape:.2f}%")

# RÂ²
ss_res = np.sum((actual - predicted) ** 2)
ss_tot = np.sum((actual - np.mean(actual)) ** 2)
r2 = 1 - (ss_res / ss_tot)
print(f"âœ… RÂ² = {r2:.4f}")

# AcurÃ¡cia Direcional
actual_direction = np.diff(actual) > 0
pred_direction = np.diff(predicted) > 0
dir_acc = np.mean(actual_direction == pred_direction) * 100
print(f"âœ… AcurÃ¡cia Direcional = {dir_acc:.1f}%")
```

## ğŸ“Š Guia de InterpretaÃ§Ã£o

### MAPE (mais importante para previsÃµes financeiras)

| MAPE | Qualidade |
|------|-----------|
| < 5% | ğŸ† Excelente |
| 5-10% | âœ… Bom |
| 10-20% | ğŸ“Š AceitÃ¡vel |
| > 20% | âš ï¸ Precisa melhoria |

### RÂ² (Coeficiente de DeterminaÃ§Ã£o)

| RÂ² | Significado |
|----|-------------|
| > 0.8 | ğŸ† Modelo explica muito bem |
| 0.5-0.8 | âœ… Bom ajuste |
| 0-0.5 | ğŸ“Š Ajuste moderado |
| < 0 | âš ï¸ Modelo pior que mÃ©dia |

### AcurÃ¡cia Direcional

| AcurÃ¡cia | Significado |
|----------|-------------|
| > 55% | âœ… Melhor que random |
| ~50% | ğŸ“Š Igual a random |
| < 50% | âš ï¸ Pior que random |
